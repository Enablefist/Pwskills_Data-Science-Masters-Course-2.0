Q1. Explain why we have to use the Exception class while creating a Custom Exception.

```python
#Q1. Explain why we have to use the Exception class while creating a Custom Exception
#Ans: We have to use the Exception class while creating a Custom 
#Exception because it is the base class for all the exceptions.
#Q2. Create a custom exception to check the validity of password entered by the user.
#Ans:
class Error(Exception):
    pass
class PasswordError(Error):
    pass
while True:
    try:
        password = input("Enter the password: ")
        if len(password) < 8:
            raise PasswordError
        elif len(password) > 15:
            raise PasswordError
        elif not any(char.isdigit() for char in password):
            raise PasswordError
        elif not any(char.isupper() for char in password):
            raise PasswordError
        elif not any(char.islower() for char in password):
            raise PasswordError
        elif not any(char in ['@','#','$'] for char in password):
            raise PasswordError
        else:
            print("Password is valid")
            break
    except PasswordError:
        print("Invalid Password")
        print("Password must be between 8 to 15 characters which contain at least one lowercase letter, one uppercase letter, one numeric digit, and one special character")
        break
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
#Ans: ArithmeticError is the base class for those built-in exceptions that are 
#raised for various arithmetic errors: OverflowError, ZeroDivisionError, FloatingPointError.
#Example:
#1. OverflowError: Raised when the result of an arithmetic operation is too large to be represented. 
#This cannot occur for integers (which would rather raise MemoryError than give up).
#Example:
import math
try:
    print(math.exp(1000))
except OverflowError:
    print("Overflow Error")
#2. ZeroDivisionError: Raised when the second argument of a division or modulo operation is zero. 
# The associated value is a string indicating the type of the operands and the operation.
#Example:
try:
    print(1/0)
except ZeroDivisionError:
    print("Zero Division Error")
#3. Floatingpoint Error:
#Raised when a floating point operation fails. This exception is always defined,
#  but can only be raised when Python is configured with the --with-fpectl option, or the WANT_SIGFPE_HANDLER symbol is defined in the pyconfig.h file.
#Example:
try:
    import decimal
    decimal.Decimal('1.00') / decimal.Decimal('0.00')
except decimal.DecimalException:
    print("Floating Point Error")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
#Ans: LookupError is the base class for the exceptions that are raised when a key or index 
#used on a mapping or sequence is invalid: IndexError, KeyError.
#Example:
#1. KeyError: Raised when a mapping (dictionary) key is not found in the set of existing keys.
#Example:
d = {1: "one", 2: "two", 3: "three"}
try:
    print(d[4])
except KeyError:
    print("Key Error")
#2. IndexError: Raised when a sequence subscript is out of range.
#Example:
l = [1,2,3,4,5]
try:
    print(l[5])
except IndexError:
    print("Index Error")
```

```python
#Q5. Explain ImportError. What is ModuleNotFoundError?
#Ans: ImportError is raised when the import statement has troubles trying to load a module. It can be raised by the import statement itself, but also by any call to __import__(), the built-in function that imports a module.
#Example:
try:
    import module
except ImportError:
    print("Import Error")
#ModuleNotFoundError is a subclass of ImportError. It is raised when a module could not be found.
#Example:
try:
    import module
except ModuleNotFoundError:
    print("Module Not Found Error")
```

Q6. List down some best practices for exception handling in python.

```python
#Q6. List down some best practices for exception handling in python.
#Ans:
#1. Use the try-except-else-finally block to handle exceptions.
#2. Use the finally block to release external resources (such as files or network connections) or 
#to perform cleanup, such as releasing memory.
#3. Use the else block to put code that should execute only if the try 
#block ran successfully and no exceptions were raised.
#4. Use the except block to catch specific exceptions and to perform specific exception handling.
#5. Use the raise statement to raise an exception explicitly.
#6. Use the assert statement to test for conditions that should never occur, such as an invalid value or a bad state.
#7. Use the with statement to simplify exception handling around calls to external resources
```
