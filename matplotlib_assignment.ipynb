Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt 
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y)
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1= np.array([0, 100, 200, 300, 400, 500])
x2=np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3= np.array([10, 20, 30, 40, 50, 60])
x4= np.array([0, 1, 2, 3, 4, 5]) 
y4= np.array([200, 350, 250, 550, 450, 150])
"""  Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data: """
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x1, y1)
ax[0, 0].set_title('subplot(2,2,1)')
ax[0, 1].plot(x2, y2)
ax[0, 1].set_title('subplot(2,2,2)')
ax[1, 0].plot(x3, y3)
ax[1, 0].set_title('subplot(2,2,3)')
ax[1, 1].plot(x4, y4)
ax[1, 1].set_title('subplot(2,2,4)')
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

```python
""" What is a bar plot?"""
# A bar chart or bar graph is a chart or graph that presents categorical 
# data with rectangular bars with heights or lengths proportional to the values that they represent.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
import matplotlib.pyplot as plt
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profit")
plt.show()

import matplotlib.pyplot as plt
plt.barh(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profit")

```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
""" What is a box plot? Why is it used? """
# A box plot is a method for graphically depicting groups of numerical data through their quartiles.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2])
plt.show() 
```
